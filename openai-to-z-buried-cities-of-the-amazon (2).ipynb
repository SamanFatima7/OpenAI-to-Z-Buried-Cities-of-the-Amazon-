This simple Python simulation mimics how buried structures (like trenches or walls) can affect terrain elevation and slope ‚Äî which are key signals in LIDAR or DEM-based archaeological surveys. It creates a central mound and embeds an artificial trench, then calculates slope and generates an anomaly map similar to real Arc Neo detections.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Step 1: Create grid and center point
width, height = 600, 600
x, y = np.meshgrid(np.arange(width), np.arange(height))
x_center, y_center = width // 2, height // 2
distance = np.sqrt((x - x_center)**2 + (y - y_center)**2)

# Step 2: Build a mound and add a trench-like anomaly
elevation = np.exp(-distance / 70.0) * 100  # radial hill
structure = (abs(x - x_center) < 30) & (abs(y - y_center) < 80)  # vertical trench zone
elevation[structure] -= 20  # depress trench region

# Step 3: Compute slope (like a LIDAR gradient response)
dx, dy = np.gradient(elevation)
slope = np.sqrt(dx**2 + dy**2)

# Step 4: Generate final anomaly map by enhancing slope contrasts
anomaly_map = elevation - slope * 2
anomaly_map = np.clip(anomaly_map, 0, 255)

# Step 5: Display the anomaly map
plt.imshow(anomaly_map, cmap='gray')
plt.title("Arc Neo LIDAR-style Anomaly Map")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy import ndimage

# 1. Simulate DEM and embed circular anomaly
size = 512
cx = cy = size // 2
y, x = np.indices((size, size))
r = np.sqrt((x - cx)**2 + (y - cy)**2)
dem = np.exp(-(r / 100)**2) * 200  # conical mound
mask = r < 60
dem[mask] -= r[mask] / 3           # simulate trench

# 2. Display DEM
plt.figure(figsize=(4, 4))
plt.imshow(dem, cmap='terrain')
plt.title("DEM (Phantom Cone)")
plt.axis('off')
plt.show()

# 3. Detect circles using Hough Transform with preprocessing + parameter tuning
dem_blur = cv2.GaussianBlur(cv2.convertScaleAbs(dem, alpha=0.5), (7, 7), 1.5)
edges = cv2.Canny(dem_blur, 30, 100)  # clean edge image

circles = cv2.HoughCircles(
    edges,
    cv2.HOUGH_GRADIENT,
    dp=1.2,
    minDist=100,
    param1=50,   # Canny high threshold
    param2=25,   # lower makes detection more permissive :contentReference[oaicite:1]{index=1}
    minRadius=40,
    maxRadius=80
)

# 4. Overlay detected circles
fig, ax = plt.subplots(figsize=(4, 4))
ax.imshow(dem, cmap='terrain')

if circles is not None:
    circles = np.round(circles[0]).astype(int)
    for x0, y0, rad in circles:
        ax.add_patch(plt.Circle((x0, y0), rad, color='red', fill=False, lw=2))
    ax.set_title("Detected Circular Anomaly")
else:
    ax.set_title("‚ö†Ô∏è No Circles Detected ‚Äî Try Tuning Parameters")
ax.axis('off')
plt.show()

# 5. Compute elevation gradient
sx = ndimage.sobel(dem, axis=1)
sy = ndimage.sobel(dem, axis=0)
grad = np.hypot(sx, sy)

plt.figure(figsize=(4, 4))
plt.imshow(grad, cmap='inferno')
plt.title("Elevation Gradient (Sobel)")
plt.axis('off')
plt.show()

# 6. Elevation profile across center
profile = dem[cy, :]
plt.figure(figsize=(5, 2))
plt.plot(profile)
plt.title("Elevation Profile Across Center")
plt.xlabel("Distance (px)")
plt.ylabel("Elevation")
plt.grid(True)
plt.tight_layout()
plt.show()

# 7. Simulated NDVI & thermal layers
ndvi = np.full((size, size), 0.8)
ndvi[mask] = 0.2
thermal = np.full((size, size), 0.65)
thermal[mask] = 0.4

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
ax1.imshow(ndvi, cmap='Greens')
ax1.set_title("Simulated NDVI")
ax1.axis('off')
ax2.imshow(thermal, cmap='coolwarm')
ax2.set_title("Simulated Thermal")
ax2.axis('off')
plt.tight_layout()
plt.show()
```
